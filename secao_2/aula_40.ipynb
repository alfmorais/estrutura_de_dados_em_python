Expressões Regulares

- Busca por padrões em texto em variaveis strings
- Palavras que tenha um padrão definido (começa com maiuscula ou minuscula)
- Palavras que começa com 'a' e termina com 'e'
- Procurar e-mails, números de telefones e CEPs
- Padrões de busca são praticamente ilimitadas

Exemplo de Frase:
"O mundo está mudando. Tecnologias utilizadas na atualidade estão tornando as tarefas cada vez mais mecanizadas e práticas, onde a mente humana é deixada livre para exercer o papel criativo"

Buscar por padrões definidos no texto:

1. Extrair todas as palavras que começam com vogais e terminam com consoantes
2. Extrair todas as palavras que possuem de 5 a 9 letras
3. Extrair todos os trechos separados por ponto (".")
4. Extrair todas as palavras que começam com "c", possuem "r", "e", ou "ou" no meio e terminam com "a", "i", ou "x". 

expressões regulares 3 funções úteis para encontrar os padrões: 

1. Método _search()_: Encontrar as posições de padrões dentro de uma string, se estes estiverem presentes. 
2. Método _match()_: Encontrar se o começo de uma string é igual a um determinado padrão.
3. Método _findall()_: Encontrar todas as substrings em uma string que correspondam a um padrão. 


- (.)  _Qualquer caractere (exceto linha nova)_
- (\w) _Qualquer caractere alfanumérico_
- (\W) _Qualquer caractere não-alfanumérico_
- (\d) _Qualquer caractere que seja digito (0-9)_
- (\D) _Qualquer caractere não digito_
- (\s) _Espaço em braco_
- (^)  _começa com_
- ($) _termina com_
- ("\\") _usado antes de metacaracteres para especificar seu significado literal_
